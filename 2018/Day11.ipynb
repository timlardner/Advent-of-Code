```python
def get_power_level(x, y, serial=8):
    rack_id = x+10
    power_level = (rack_id * y + serial) * rack_id
    hundreds_digit = int(str(power_level)[-3:-2] or 0)
    return hundreds_digit - 5
```

```python
from collections import defaultdict

import numpy as np

# SERIAL = 18
SERIAL = 9110

def build_cache(serial):
    #cache = defaultdict(dict)
    cache = np.zeros((300,300))
    for x in range(300):
        for y in range(300):
            cache[x][y] = get_power_level(x+1, y+1, serial=serial)
    return cache

score_cache = {}

cache = build_cache(SERIAL)


max_score = 0
best_coords = None
max_len = 300
for x_start in range(1,max_len+1):
    print('Testing start x co-ordinate: ', x_start)
    for y_start in range(1,max_len+1):
        for size in range(1, max_len+1):
            if size == 1:
                score = cache[x_start-1,y_start-1]
            else:
                if x_start + size > 300 or y_start + size > 300:
                    break
                    
                score = score_cache[(x_start,y_start,size-1)]
                        
                # Get edge scores
                y = y_start -1 + size
                for x in range(x_start, x_start+size):
                    score += cache[x-1][y-1]

                x = x_start -1 + size
                for y in range(y_start, y_start+size-1):  # -1 because we only add the corner once
                    score += cache[x-1,y-1]
                    
            
            score_cache[(x_start,y_start,size)] = score
            if score > max_score:
                print('Start of {}, {} with size of {} has score {}'.format(x_start,y_start,size, score))
                max_score = score
                best_coords = (x_start,y_start,size)

print(best_coords)
```
