```python
def get_power_level(x, y, serial=8):
    rack_id = x+10
    power_level = (rack_id * y + serial) * rack_id
    hundreds_digit = int(str(power_level)[-3:-2] or 0)
    return hundreds_digit - 5
```

```python
from collections import defaultdict

def build_cache():
    cache = defaultdict(dict)
    for x in range(1,301):
        for y in range(1,301):
            cache[x][y] = get_power_level(x, y, serial=9110)
    return cache

score_cache = {}

cache = build_cache()


max_score = 0
best_coords = None
max_len = 300
for x_start in range(1,max_len+1):
    print('Testing start position ', x_start)
    for y_start in range(1,max_len+1):
        #for size in range(1,max_len+1-max((y_start, x_start))):  # 4 for part 1
        for size in range(1,max_len):  # 4 for part 1
            if size == 1:
                score = cache[x_start][y_start]
            else:
                if x_start + size > 300 or y_start + size > 300:
                    continue
                    
                #print('Testing start of {},{} with size of {}'.format(x_start,y_start,size))
                
                score = score_cache[(x_start,y_start,size-1)]
                        
                # Get edge scores
                y = y_start -1 + size
                for x in range(1, x_start+size):
                    #print('Adding {},{} to result'.format(x,y))
                    score += cache[x][y]

                x = x_start -1 + size
                for y in range(1, y_start+size-1):  # -1 because we only add the corner once
                    #print('Adding {},{} to result'.format(x,y))
                    score += cache[x][y]
            
            
                    
            score_cache[(x_start,y_start,size)] = score
            if score > max_score:
                max_score = score
                best_coords = (x_start,y_start,size)
print(best_coords)
```
